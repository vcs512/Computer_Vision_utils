# Imports

```python
import cv2 as cv
import sys
import matplotlib.pyplot as plt

print("OpenCV version = ", cv.__version__)
```

# Load image

```python
img = cv.imread(filename="example.jpg",
                flags=cv.IMREAD_COLOR)
```

# Convert images

```python
img_RGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
print("RGB shape = ", img_RGB.shape)

img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
print("Gray shape = ", img_gray.shape)
```

## Size

```python
img_reduced = cv.resize(src=img_gray,
                        dsize=(50,50))
print("Reduced shape = ", img_reduced.shape)

img_expanded = cv.resize(src=img_gray,
                        dsize=(2*img_gray.shape[0], 2*img_gray.shape[1]))
print("Expanded shape = ", img_expanded.shape)
```

# Plot images

```python
figure = plt.figure(1, figsize=(10,30))

plt.subplot(131)
plt.imshow(img)
plt.title("Standard BGR")

plt.subplot(132)
plt.imshow(img_RGB)
plt.title("RGB converted")

plt.subplot(133)
plt.imshow(img_gray, cmap="gray")
plt.title("Grayscale")

plt.show()
```

```python
sizes = [img_gray, img_reduced, img_expanded]

figure = plt.figure(2, figsize=(10,30))
for image in range(len(sizes)):
    plt.subplot(131 + image)
    plt.imshow(sizes[image], cmap="gray")
    

plt.show()
```

# Save images

```python
cv.imwrite(filename="./gray.png",
           img=img_gray)
```
